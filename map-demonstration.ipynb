```python
# Import python libraries
import pandas as pd
import hvplot.pandas
import holoviews as hv
import osmnx as osm
```

```python
# Search for boundary using OSM

cu_boundary = osm.geocode_to_gdf(
                        "W46226108",
                        by_osmid=True
)

cu_boundary
```

```python
cu_boundary.plot()
```

```python
cu_map = cu_boundary.hvplot(
    title = "Map of CU Boulder main campus",
    geo = True, tiles = 'EsriImagery',

    fill_color = 'white',fill_alpha = 0.2,
    line_color = 'skyblue', line_width =5,
    frame_width = 400, frame_height = 400)
hv.save(cu_map,'cu.html')

cu_map
```

```python
uof_sl = osm.features_from_address(
    'University of Peradeniya',
    {'amenity':['university']})

uof_sl

```

```python
uof_map = uof_sl.hvplot(
    title = "Map of University of Peradeniya, Sri Lanka",
    geo = True, tiles = 'EsriImagery',

    fill_color = 'white',fill_alpha = 0.2,
    line_color = 'skyblue', line_width =5,
    frame_width = 400, frame_height = 400)
hv.save(uof_map,'uof.html')

uof_map
```
